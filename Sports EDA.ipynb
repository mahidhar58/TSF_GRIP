```python
# Importing the required libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Retrieving the data

```python
deliveries = pd.read_csv('./data/deliveries.csv')
deliveries.head()
```

```python
matches = pd.read_csv('./data/matches.csv')
matches.head()
```

Checking for null values in both datasets

```python
matches.isna().any()
```

```python
deliveries.isna().any()
```

```python
matches.info()
```

```python
deliveries.info()
```

```python
# Dropping umpire columns, because we don't need those

matches = matches.drop(['umpire1','umpire2','umpire3'],axis=1)
```

```python
matches.head(1)
```

# Preparing the data for plotting

```python
# List of all teams participated

teams = matches['winner'].unique()
teams
```

```python
teams = [i for i in teams if pd.isnull(i)==False]
teams
```

```python
# Number of wins for all teams stored in dictionary

wins = dict()
for team in teams:
    wins[team] = len([i for i in matches['winner']==team if i==True])
wins
```

```python
# Converting the dictionary into DataFrame

wins = dict(sorted(wins.items(),reverse=True,key= lambda x:x[1]))
total_wins = pd.DataFrame()
total_wins.insert(0,'team','')
total_wins.insert(1,'wins','')
for i in wins:
    total_wins.loc[len(total_wins)] = [i,wins[i]]
total_wins
```

```python
# List of all batsmen

batsmen = deliveries['batsman'].unique()
batsmen
```

```python
# Calculating the number of runs scored by every batsman and storing in dictionary

most_runs = dict()
for batsman in batsmen:
    most_runs[batsman] = sum((deliveries.loc[deliveries['batsman']==batsman])['batsman_runs'])
most_runs
```

```python
most_runs = sorted(most_runs.items(),reverse=True,key= lambda x:x[1])
```

```python
most_runs
```

```python
# Converting the dictionary into DataFrame

mostRuns = pd.DataFrame(most_runs)
mostRuns.columns = ['players','runs']
```

```python
toss_factor = []
tot = len(matches)
for i in range(tot):
    if(matches.loc[i]['toss_winner']==matches.loc[i]['winner']):
        toss_factor.append(matches.loc[i]['winner'])
toss_factor
len(toss_factor)
```

```python
tot-len(toss_factor)
```

```python
# Filling the null values

null_city=[]
for index,boolean in enumerate(matches['city'].isna()):
    if boolean:
        null_city.append([index,matches.loc[index]['venue']])
null_city
```

Since the venues are in Dubai, we can fill the null values with "Dubai"

```python
matches.city = matches.city.fillna('Dubai')
```

```python
matches.city.isna().any()
```

```python
# List of all the bowlers

bowlers = deliveries['bowler'].unique()
bowlers
```

```python
# Calculating the number of wickets taken by each bowler and storing in a dictionary

wickets = []
dismissals = deliveries['player_dismissed']
for index,player in enumerate(dismissals):
    if type(player)==str:
        wickets.append(deliveries.loc[index]['bowler'])
wickets
most_wickets = dict()
for bowler in bowlers:
    most_wickets[bowler] = wickets.count(bowler)
most_wickets
```

```python
# Converting from dictionary to DataFrame

most_wickets = dict(sorted(most_wickets.items(), key = lambda x: x[1], reverse=True))
total_wickets = pd.DataFrame()
total_wickets.insert(0,'bowler','')
total_wickets.insert(1,'wickets','')
for i in most_wickets:
    total_wickets.loc[len(total_wickets)] = [i,most_wickets[i]]
total_wickets.head()
```

```python
# List of all players winning Player of Match

MoM_list = matches['player_of_match']
MoM_list
```

```python
MoM = MoM_list.unique()
MoM
```

```python
MoM = list(MoM)
MoM_list = list(MoM_list)
```

```python
# Calculating number of awards each player won

most_MoM = dict()
for player in MoM:
    most_MoM[player] = MoM_list.count(player)
most_MoM
```

```python
# Converting dictionary into DataFrame

most_MoM = dict(sorted(most_MoM.items(),key= lambda x: x[1], reverse=True))
total_awards = pd.DataFrame()
total_awards.insert(0,'player','')
total_awards.insert(1,'awards','')
for i in most_MoM:
    total_awards.loc[len(total_awards)] = [i,most_MoM[i]]
total_awards
```

```python
# Sixes hit by a player

sixes = []
for index,runs in enumerate(deliveries['batsman_runs']):
    if(runs == 6):
        sixes.append(deliveries.loc[index]['batsman'])
sixes
```

```python
# Count of sixes hit by every player

most_sixes = dict()
for batsman in batsmen:
    most_sixes[batsman] = sixes.count(batsman)
most_sixes
```

```python
# Converting dictionary into DataFrame

most_sixes = dict(sorted(most_sixes.items(),key= lambda x: x[1],reverse=True))
total_sixes = pd.DataFrame()
total_sixes.insert(0,'player','')
total_sixes.insert(1,'sixes','')
for i in most_sixes:
    total_sixes.loc[len(total_sixes)] = [i,most_sixes[i]]
total_sixes
```

```python
# Fours hit by players

fours = []
for index,runs in enumerate(deliveries['batsman_runs']):
    if(runs == 4):
        fours.append(deliveries.loc[index]['batsman'])
fours
```

```python
# Count of fours hit by each player

most_fours = dict()
for batsman in batsmen:
    most_fours[batsman] = fours.count(batsman)
most_fours
```

```python
# Converting dictionary into DataFrame

most_fours = dict(sorted(most_fours.items(),key= lambda x: x[1],reverse=True))
total_fours = pd.DataFrame()
total_fours.insert(0,'player','')
total_fours.insert(1,'fours','')
for i in most_fours:
    total_fours.loc[len(total_fours)] = [i,most_fours[i]]
total_fours
```
